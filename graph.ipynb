# Loss of resnet18 and resnet34 on kdd17

```python
# 读取日志文件
log_files = [
    '/Users/lixi/Desktop/TB-3/Adv-ALSTM-main-2/original_code_updated/train_resnet_kdd.log',
    '/Users/lixi/Desktop/TB-3/Adv-ALSTM-main-2/original_code_updated/train_resnet34_lstm_kdd.log',
    # 添加更多文件路径...
]
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple', 'tab:pink']
labels = ['Test Loss of ResNet18-LSTM', 'Test Loss of ResNet34-LSTM']

# 遍历每个日志文件并提取数据
for i, log_file in enumerate(log_files):
    epochs = []
    test_accuracies = []
    test_losses = []
    test_mccs = []
    reading_data = False

    with open(log_file, 'r', encoding='utf-8') as file:
        for line in file:
            if line.startswith('----->>>>> Training:'):
                reading_data = True
            if reading_data:
                if line.startswith('	Test per:'):
                    parts = line.split()
                    test_accuracy = parts[3].strip(',')
                    test_loss = parts[-1]
                    test_mcc = parts[5].strip('}')
                    test_accuracies.append(float(test_accuracy))
                    test_losses.append(float(test_loss))
                    test_mccs.append(float(test_mcc))
        # 创建一个表示 epoch 的列表
        epochs = list(range(1, len(test_accuracies) + 1))

    # 绘制测试精度曲线，每个文件对应一条曲线
    
    plt.plot(epochs, test_losses, label=labels[i], color=colors[i+3])
    #labels.append(f'Test Acc {i + 1}')

# 添加标签和标题
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Test Loss Comparison')
plt.legend(labels)

# 保存图像
plt.savefig('Test_Loss_Comparison_kdd.svg', format='svg', dpi=1200)

# 显示图像
plt.show()

```
